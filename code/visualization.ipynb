```python
# Visualization
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import os
import pandas as pd

plt.rc('xtick', labelsize=15)     
plt.rc('ytick', labelsize=15)

```

```python
### Set your path to the folder containing the .csv files
PATH = 'results/accuracy/' # Use your path

### Fetch all files in path
fileNames = os.listdir(PATH)

### Filter file name list for files ending with .csv
fileNames = [file for file in fileNames if '.csv' in file]

fig=plt.figure(figsize=(16, 9), dpi= 80, facecolor='w', edgecolor='k')

# define number of ticks
plt.locator_params(nbins=15)
i = 0
### Loop over all files
for file in fileNames:
    
    ### Read .csv file and append to list
    df = pd.read_csv(PATH + file, index_col = 0, usecols = ['Step','Value'])
    ### Create line for every file
    label_ = fileNames[i].replace('.csv', '')
    if label_ == 'DenseNet169':
        plt.plot(df, label=label_, marker='*', ms=10)
    elif label_ == 'NASNetLarge':
        plt.plot(df, label=label_, marker='o', ms=10)
    else:
        plt.plot(df, label=label_)
    legend = plt.legend(loc='lower right', shadow=True, fontsize='xx-large')
    plt.xlim(0, 29)
    i = i + 1
i = 0

plt.ylabel('Validation Accuracy', fontsize=15)
plt.xlabel('Value', fontsize=15)
```
